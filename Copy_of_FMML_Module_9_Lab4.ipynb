## Using learnt representations
In practice, very few people train an entire Convolutional Network from scratch (with random initialization), because it is relatively rare to have a dataset of sufficient size. Instead, it is common to pretrain a ConvNet on a very large dataset (e.g. ImageNet, which contains 1.2 million images with 1000 categories), and then use the ConvNet either as an initialization or a fixed feature extractor for the task of interest.


We'll train a model to classify ants and bees. We have about 120 training images each for ants and bees. There are 75 validation images for each class. Usually, this is a very small dataset to generalize upon, if trained from scratch. Since we are using transfer learning, we should be able to generalize reasonably well.

This dataset is a very small subset of imagenet.




```python
# Import packages
import torch
import torchvision
import torchvision.transforms as transforms
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader, SubsetRandomSampler
import torch.nn as nn
from torch import optim
import torch.nn.functional as F

import matplotlib.pyplot as plt
import numpy as np
import time
import os
```

```python
# Device configuration (whether to run on GPU or CPU)
device = 'cpu' #torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
# Download and extract dataset
!wget https://download.pytorch.org/tutorial/hymenoptera_data.zip
!unzip -q hymenoptera_data.zip
```

```python
# Data augmentation and normalization for training
# Just normalization for validation
train_transform = transforms.Compose([
        transforms.RandomResizedCrop(224),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])
val_transform = transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

data_dir = './hymenoptera_data'
train_dataset = ImageFolder(os.path.join(data_dir, 'train'), train_transform)
val_dataset = ImageFolder(os.path.join(data_dir, 'val'), val_transform)
train_dataloader = DataLoader(train_dataset, batch_size=4,shuffle=True, num_workers=2)
val_dataloader = DataLoader(val_dataset, batch_size=4,shuffle=True, num_workers=2)
class_names = train_dataset.classes
```

```python
def imshow(inp, title=None):
    """Imshow for Tensor."""
    inp = inp.numpy().transpose((1, 2, 0))
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    inp = std * inp + mean
    inp = np.clip(inp, 0, 1)
    plt.imshow(inp)
    if title is not None:
        plt.title(title)
    plt.pause(0.001)  # pause a bit so that plots are updated


# Get a batch of training data
inputs, classes = next(iter(train_dataloader))

# Make a grid from batch
out = torchvision.utils.make_grid(inputs)

imshow(out, title=[class_names[x] for x in classes])
```

```python
def train_model(model, train_dataloader, criterion, optimizer, num_epochs=25):

    for epoch in range(num_epochs):
        print(f'Epoch {epoch}/{num_epochs - 1}')
        print('-' * 10)
        model.train()
        running_loss = 0.0
        running_corrects = 0

        # Iterate over data.
        for i, (inputs, labels) in enumerate(train_dataloader):
            inputs = inputs.to(device)
            labels = labels.to(device)

            # zero the parameter gradients
            optimizer.zero_grad()

            # forward
            # track history if only in train
            with torch.set_grad_enabled(True):
              outputs = model(inputs)
              _, preds = torch.max(outputs, 1)
              loss = criterion(outputs, labels)

              # backward + optimize only if in training phase
              loss.backward()
              optimizer.step()

            # statistics
            running_loss += loss.item() * inputs.size(0)
            running_corrects += torch.sum(preds == labels.data)

        epoch_loss = running_loss / len(train_dataloader.dataset)
        epoch_acc = running_corrects.double() / len(train_dataloader.dataset)

        print(f'Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}')

        print()
```

```python
def visualize_model(model, num_images=10):
    was_training = model.training
    model.eval()
    images_so_far = 0
    fig = plt.figure()

    with torch.no_grad():
        for i, (inputs, labels) in enumerate(val_dataloader):
            inputs = inputs.to(device)
            labels = labels.to(device)

            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)

            for j in range(inputs.size()[0]):
                images_so_far += 1
                ax = plt.subplot(num_images//2, 2, images_so_far)
                ax.axis('off')
                ax.set_title(f'predicted: {class_names[preds[j]]}')
                imshow(inputs.cpu().data[j])

                if images_so_far == num_images:
                    model.train(mode=was_training)
                    return
        model.train(mode=was_training)
```

```python
from torchvision import models

model_ft = models.resnet18(pretrained=True)
num_ftrs = model_ft.fc.in_features
# Here the size of each output sample is set to 2.
# Alternatively, it can be generalized to nn.Linear(num_ftrs, len(class_names)).
model_ft.fc = nn.Linear(num_ftrs, 2)

model_ft = model_ft.to(device)

loss_func = nn.CrossEntropyLoss()

# Observe that all parameters are being optimized
optimizer = optim.SGD(model_ft.parameters(), lr=0.001, momentum=0.9)
```

```python
num_epochs = 5
train_model(model_ft, train_dataloader, loss_func, optimizer, num_epochs)
```

```python
visualize_model(model_ft)
```

#1ST QUESTION ANSWER

Data augmentation techniques, such as resizing, cropping, rotating, flipping, and adding noise to training data, play a crucial role in machine learning and deep learning tasks, especially in computer vision. Here are some key reasons why data augmentation is significant:

Increased Dataset Size: Data augmentation allows you to artificially increase the size of your training dataset by applying transformations to existing data. This can help prevent overfitting and improve the generalization of your model.

Regularization: By augmenting the training data with variations of the original samples, you effectively introduce regularization to the model. Regularization helps prevent the model from memorizing the training data and encourages it to learn more robust and generalizable features.

Improved Generalization: Data augmentation exposes the model to a wider variety of data during training, making it more robust to variations and distortions present in real-world data. This can lead to better performance on unseen data.

Invariant Representation Learning: Data augmentation helps the model learn features that are invariant to certain transformations. For example, if you rotate or flip an image, objects in the image remain the same despite the transformation. By exposing the model to these variations during training, it learns to focus on the invariant features, which can improve performance on tasks where such variations are present.

Reduced Memorization: Without data augmentation, a model might memorize specific details of the training data, leading to poor generalization on unseen data. Data augmentation disrupts this memorization process by presenting variations of the same data, encouraging the model to learn more abstract and invariant representations.

Robustness to Variations: Real-world data often contains variations such as changes in lighting conditions, occlusions, and deformations. By augmenting the training data with these variations, the model becomes more robust and better equipped to handle such variations during inference.

Overall, data augmentation is a powerful technique for improving the performance, generalization, and robustness of machine learning models, particularly in tasks such as image classification, object detection, and segmentation.

#2ND QUESTION ANSWER

If you don't use a pretrained ResNet model and train the model from scratch (i.e., with pretrained=False), the performance might not be as good as when using transfer learning with a pretrained model. Training a deep neural network from scratch on a small dataset like the one you described (ants vs. bees) can lead to overfitting due to limited data.

Without the benefits of transfer learning from a pretrained model, the network has to learn features from scratch, which can be challenging with limited data. As a result, the model might struggle to generalize well to unseen data and may not achieve optimal performance.

However, training without pretraining allows the model to learn features specifically tailored to the task at hand, which could potentially yield better results if the dataset is large enough and representative of the target task. Nonetheless, in most cases, transfer learning with a pretrained model tends to provide better results, especially when dealing with small datasets like the one you have.







#3RD QUESTION

Yes, even if the ResNet model was pretrained on a dataset significantly different from the ants vs. bees data, you would likely still get good performance by using this pretrained model. This is one of the main advantages of transfer learning.

The pretrained ResNet model has already learned generic features from a large and diverse dataset such as ImageNet, which contains a wide variety of object categories. These learned features capture general patterns and representations that are useful for a variety of visual recognition tasks.

When fine-tuning the pretrained ResNet model on a specific task like classifying ants vs. bees, the model adapts its learned features to better suit the characteristics of the new dataset. While some of the features learned from ImageNet may not directly correspond to ants or bees, many lower-level features such as edges, textures, and colors are likely to be transferable and useful for the new task.

Furthermore, the deeper layers of the ResNet model capture increasingly abstract and high-level features, which are often more task-agnostic and transferable across datasets. By fine-tuning these layers while keeping the early layers frozen (or using them as fixed feature extractors), the model can effectively leverage the knowledge gained from the pretrained model to improve its performance on the new task.

Overall, using a pretrained ResNet model, even if pretrained on a different dataset, can still lead to good performance on the ants vs. bees classification task, especially considering the small size of the dataset available for training.
